```python
import matplotlib.pyplot as plt
from desi_retriever.blanc import get_specs, get_rvspec_models
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
# fetch spectra for a given tile, night, fiber and targetid
# Return multiple sprectra
D = get_specs(tileid=63075, night=20200220, fiber=0, targetid=35190987189916575)
```

```python
plt.clf()
plt.subplot(311)
plt.plot(D[0]['b_wavelength'], D[0]['b_flux'], color='blue')
plt.subplot(312)
plt.plot(D[0]['r_wavelength'], D[0]['r_flux'], color='green')
plt.subplot(313)
plt.plot(D[0]['z_wavelength'], D[0]['z_flux'], color='red')
```

```python
# multiple exposures
plt.plot(D[0]['b_wavelength'], D[0]['b_flux'], color='blue')
plt.plot(D[1]['b_wavelength'], D[1]['b_flux'], color='green')
```

```python
# Get the rvspecfit models
D = get_specs(tileid=63075, night=20200220, fiber=0, targetid=35190987189916575,coadd=True)
Dmod = get_rvspec_models(
    tileid=63075, night=20200220, fiber=0, targetid=35190987189916575,coadd=True)
```

```python
plt.subplot(311);
plt.plot(D[0]['b_wavelength'],D[0]['b_flux'],color='lightblue');
plt.plot(Dmod[0]['b_wavelength'],Dmod[0]['b_model'],color='blue');
plt.subplot(312);
plt.plot(D[0]['r_wavelength'],D[0]['r_flux'],color='lightgreen');
plt.plot(Dmod[0]['r_wavelength'],Dmod[0]['r_model'],color='green');
plt.subplot(313);plt.plot(D[0]['z_wavelength'],D[0]['z_flux'],color='red',alpha=.5);
plt.plot(Dmod[0]['z_wavelength'],Dmod[0]['z_model'],color='red');
```
